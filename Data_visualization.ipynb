# Target plot
## A detailed plot for the target to get a clear idea of it's distribution

```python
D_cols = data1.columns.to_list()

import seaborn as sns
sns.set(context='notebook', style='darkgrid', palette='deep', font='sans-serif', font_scale=1, color_codes=False, rc=None)

total = float(len(data1[D_cols[0]]))
plt.figure(figsize =(6,6))
sns.set(style="darkgrid")
ax = sns.countplot(data1[D_cols[0]])
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2., height + 3, '{:1.2f}'.format(height/total), ha = "center")
plt.title("Target Plot", fontsize = 20)
plt.show()
```

# Univariate plots

```python
for col in D_cols[1:]:
    plt.figure(figsize=(10,4))
    sns.countplot(x=col, data=data1, palette='icefire')
    plt.title(col, fontsize=14)
    plt.show()
    print("% of total:")
    print(round(data1[col].value_counts()/data1.shape[0], 4)*100)
```
